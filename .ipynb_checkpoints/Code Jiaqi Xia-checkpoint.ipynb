### I.3) Main challenges

We need to create relevant indicators that quantify market sentiment with an ecological focus. 

Central bank roadmaps have a huge influence on the economy, while browsing the speeches of central bank members, I noticed that in the titles of the first 25 speeches, 5 directly mention the issues of sustainable development. I therefore found it relevant to create an indicator based on the occurrences of speeches evoking sustainable development. 

For this we need a metric that measures how much a speech evokes sustainable economy. One of the challenges is that commonly used text processing tools like Loughran-McDonald focus a lot on traditional finance concepts (positive, negative, gain, loss...) and little on concepts evoking ESG themes. My research into the current state of the art did not find any suitable tools for this task. We will have to be creative.

However, if these speeches can be closely followed by "insiders", they have little audience among the general public who are also significant players in the market. So in addition to the previous indicator, I considered it relevant to quantify the media coverage of global warming in the world to make it into another indicator more in line with public opinion sentiment.

The S&P clean energy index dynamics doesn't seems to follow a simple ARMA process, not even a SARIMA. We can't use existing classical statistical predictors for time series. And there could be longterm dependances between S&P index to past ecological indicator, or to its own past.

We successfully extracted relevant elements from an online database to construct two quantitative indicators reflecting ecological-themed market sentiment.

As anticipated, our model exhibits good predictive performance on the training data, where the predicted curve closely aligns with the true curve. However, the quality of predictions is quite poor on the test data, the return and volatility curves didn't overlap well. By comparing models trained with and without our 2 indicators, we can't conclude that they improve prediction on S&P Global Clean Energy Index. Taking sub_optimal hyperparameters doesn't seem to change this conclusion.

## III) Analytics and learning strategies


### III.1) Preprocessing

First, we want to turn each data into a time serie defined every day. 

For the index, we performed linear interpolation between date where the values are not defined (days where the market is closed)

For central bankers speechees, we attribute an "ecoscore" to each central banker speech using term frequency-inverse document frequency technique (TF-IDF) with respect to a list of words related with sustainable development. The IDF are computed on speechees of training data, and the list of words are obtained on the website https://relatedwords.org/relatedto/sustainable%20development which obtained them with words embedding.
We give a importance factor of 30 to title content incomparison to text.

We can criticize the use of TF-IDF in the sense that it does not take into account the real meaning of the sentences but only the presence or absence of certain words. But in this particular case it still seems relevant because today if a central banker gives a speech, for example, "decarbonizing", it is very likely that the context would be positive toward decarbonizing  like "Investments must be oriented to decarbonize the economy" and it is unlikely that the context would be negative toward decarbonizing like "I think that the immediate profit is more important than decarbonizing the economy"

```python
#preprocessing functions
title_coef = 30
def count_texts_containing_phrase(texts, target_phrase):
    count = 0
    for text in texts:
        # Check if the target phrase appears as a whole word in the text
        if f' {target_phrase} ' in f' {text} ':
            count += 1
    return count
def compute_idf(words, sp_text, train_title, sp_date):
    words = [[word, 0, 0] for word in words]
    train_text = sp_text[sp_date < np.datetime64(test_start_date)]
    train_title = sp_title[sp_date < np.datetime64(test_start_date)]
    N = train_text.shape[0]
    n = len(words)
    for k in range(n):
        words[k][1] = count_texts_containing_phrase(train_title, words[k][0])/N
        words[k][2] = count_texts_containing_phrase(train_text, words[k][0])/N
    #words = sorted(words, key=lambda x: x[2])
    print(words[:10])
    return(words)
def compute_sp_ecoscore(sp_title, sp_text, idf):
    duration = sp_title.shape[0]
    ecoscore = np.zeros(duration)
    print(sp_title[0].count(idf[0][0])/(idf[0][1]+1))
    for k in range(duration):
        e = 0
        for i in range(len(idf)):
            e+= (sp_title[k].count(idf[i][0])*title_coef/(idf[i][1]+1) + sp_text[k].count(idf[i][0])/(idf[i][2]+1))
        ecoscore[k] = e
    plot(np.arange(0,ecoscore.shape[0],1),ecoscore,"days, day 0 : "+ str(datetime(2017, 1, 1)),"Value","ecoscore",".")
    return(ecoscore)
def filtering(ecoscore):

    data_reshaped = ecoscore.reshape(-1, 1)
    num_clusters = 3

    # Perform K-Means clustering
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(data_reshaped)

    labels = kmeans.labels_
    centers = kmeans.cluster_centers_
    lowest_cluster = np.argmin(centers)

    # Create a mask to filter data points in the lowest cluster
    mask_lowest_cluster = labels == lowest_cluster

    # Create a new array with values from the original array where the lowest cluster is set to 0
    filtered_ecoscore = np.where(mask_lowest_cluster, 0, ecoscore)
    # Plot the original data and cluster centers
    plt.scatter(np.arange(len(filtered_ecoscore)), filtered_ecoscore, c=labels, cmap='viridis', marker='o', label='Data Points')
    plt.scatter(np.arange(num_clusters), centers, c='red', marker='x', label='Cluster Centers')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.legend()
    plt.show()
    return(filtered_ecoscore)
        
def place_ecoscore_on_timeline(ecoscore, sp_date):
    
    ecoscore_timeline = np.zeros(total_duration)
    
    for i in range(sp_date.shape[0]):
        k = (sp_date[i]-np.datetime64(start_date)).astype('timedelta64[D]').astype(int)
        
        ecoscore_timeline[k] += ecoscore[i]
    
    
    plot(np.arange(0,ecoscore_timeline.shape[0],1),ecoscore_timeline,"days","ecoscore","ecoscore_timeline","-")
    return(ecoscore_timeline)
def resonance(x, resonance_coef ,p):
    y = np.zeros(x.shape[0])
    for i in range(x.shape[0]):
        if i==0:
            past = 0
        else:
            past = y[i-1]
        y[i] = x[i] + resonance_coef*past
    if p:
        plot(np.arange(0,y.shape[0],1),y,"days","ecoscore","ecoscore timeline with resonance","-")
    return(y)
def truncate(index, ecoscore_timeline, rate, media):
    m = min([ecoscore_timeline.shape[0], media.shape[0], index.shape[0], rate.shape[0]])
    true_total_duration = m
    return(index[:m], ecoscore_timeline[:m], media[:m], rate[:m], true_total_duration)
def get_trend(index, horizon):

    returns = np.diff(index) / index[:-1]
    index_r = np.zeros((returns.shape[0] - horizon + 1))
    index_v = np.zeros((returns.shape[0] - horizon + 1))
    for i in range(returns.shape[0] - horizon + 1 ):
        window = returns[i:i+horizon]
        index_r[i] = np.mean(window)
        index_v[i] = np.std(window)

    return(index_r, index_v)
```

```python
idf = compute_idf(words, sp_text, sp_title, sp_date)
```

```python
ecoscore = compute_sp_ecoscore(sp_title, sp_text, idf)
```

We assume that in reality some speechees do not deal with the subject of sustainable development at all and have a non-zero ecoscore through pure noise, Therefore we use unsupervised learning to separate the speechees into 3 tiers according to their ecoscore (clustering) ,then we set the ecoscore of the lowest tier to 0. Finally we create a time serie of speechees by placing ecoscore on a timeline (time step = 1 day) according to their dates to have a value (could be 0) defined every day. 

```python
filtered_ecoscore = filtering(ecoscore)
```

```python
ecoscore_timeline0 = place_ecoscore_on_timeline(filtered_ecoscore, sp_date)
```

However, the first tests give results in the sense that the addition of the ecoscore reduced the prediction performance. We thought that it was because the speeches were represented by discontinuous peaks whereas in reality they have a resonance over a certain period. This makes it difficult for the neural network to model long-term causal links between future trends and a long-ago speech peak. Therefore we had the idea of doing an autoregressive transformation of on the array of ecoscore : 


$y_t = x_t + y_{t-1} * \text{resonance_coef},
\quad t > 1,  \\
\quad y_0 = x_0$

Here we choose an arbitrary resonance coef with a half life of 3 days. It doesn't have to reflect reality cognitively speaking, its juste a tool to improve learning. 

```python
resonance_coef = 0.5**(1/3)
ecoscore_timeline = resonance(ecoscore_timeline0, resonance_coef ,True)
```

### III.2 ) Structure data

For forecasting, we adopt a neural network methodology. To facilitate training, we organize the data as follows: The input to the network, denoted as X, comprises 2D arrays representing information from the last 30 days for each of the three datasets (index values, ecoscore, and media coverage), all standardized. The associated labels consist of the average return (Y_r) and volatility (Y_v) of the index over a 7-day period.


```python
def scale_and_structure(index, index_r, index_v, ecoscore_timeline, media, look_back):

    index = index.reshape(-1, 1)
    ecoscore_timeline = ecoscore_timeline.reshape(-1, 1)
    media = media.reshape(-1, 1)
    index_r = index_r.reshape(-1, 1)
    index_v = index_v.reshape(-1, 1)
    scalers = []
    scaler = StandardScaler()
    scaler.fit(index)
    scalers.append(scaler)
    index = scaler.transform(index)
    
    scaler = StandardScaler()
    scaler.fit(ecoscore_timeline)
    scalers.append(scaler)
    ecoscore_timeline = scaler.transform(ecoscore_timeline)
    
    scaler = StandardScaler()
    scaler.fit(media)
    scalers.append(scaler)
    media = scaler.transform(media)

    X0 = np.column_stack((index, ecoscore_timeline, media))
    n = index_r.shape[0]
    X, Y_r, Y_v = np.zeros((n - look_back, look_back, 3)), np.zeros((n - look_back, 1, 1)), np.zeros((n - look_back, 1, 1))
    for i in range(n - look_back):
        X[i,:,:] = X0[i:i+look_back]
        Y_r[i,:,:] = index_r[i+look_back, :]
        Y_v[i,:,:] = index_v[i+look_back, :]
    return X, Y_r, Y_v, scalers
def split_train_test(X, Y_r, Y_v):
    X_train = X[:train_duration-look_back-horizon]
    X_test = X[train_duration-look_back-horizon:]
    Y_r_train = Y_r[:train_duration-look_back-horizon]
    Y_r_test = Y_r[train_duration-look_back-horizon:]
    Y_v_train = Y_v[:train_duration-look_back-horizon]
    Y_v_test = Y_v[train_duration-look_back-horizon:]
    return(X_train, X_test, Y_r_train, Y_r_test, Y_v_train, Y_v_test)
```

```python
look_back = 30  # Number of past time steps to consider

X, Y_r, Y_v, scalers = scale_and_structure(index, index_r, index_v, ecoscore_timeline, media ,look_back)
```

### III.3) Machine learning

We chose to use LSTM, a neural network architecture which is particularly suited to capturing long-term dependencies.
A short presentation of LSTM : https://colah.github.io/posts/2015-08-Understanding-LSTMs/
We use 2 seperate networks that take the same input, the first one is trained to predict return and the second one volatility. to avoid negative or irrealistic small volatility values we added a custom activation function that is differentiable, big gradient (to avoid gradient vanishing) and has minimum a (a taken as the minimum volatility observed on the training data)

The selection of hyperparameters of each network, such as the learning rate and weight decay for the optimizers, as well as the dimensions of the LSTM and dense layers, has been the result of thorough experimentation. These choices strike a balance between creating a sufficiently complex network for the task at hand and mitigating the risk of overfitting. It's worth noting that, in isolated instances, the return and volatility predictions exhibited occasional unpredictable extreme values. To address this concern, the log_cosh loss function is employed, providing a more robust approach that avoids excessively penalizing outliers.

```python

def custom_activation(x, a):
    return a+x**2
def train(X, Y_r, Y_v, look_back, min_v, verbose):
    optimizer_r = Adam(learning_rate=0.0002, weight_decay=10**-5)
    optimizer_v = Adam(learning_rate=0.00025, weight_decay=10**-4)
    # Build the LSTM model
    model_r = Sequential()
    model_r.add(LSTM(units=100, input_shape=(look_back, 3)))
    model_r.add(Dense(units=64, activation='relu'))
    model_r.add(Dense(units=1)) 
    model_r.compile(optimizer=optimizer_r, loss='log_cosh')

    model_v = Sequential()
    model_v.add(LSTM(units=140, input_shape=(look_back, 3)))
    model_v.add(Dense(units=64, activation='relu'))
    model_v.add(Dense(units=1, activation=lambda x: custom_activation(x, min_v))) 
    model_v.compile(optimizer=optimizer_v, loss='log_cosh')
    
    # Train the model that predict returns
    model_r.fit(X, Y_r, epochs=10, batch_size=32, verbose=verbose)
    
    # Train the model that predict volatility
    model_v.fit(X, Y_v, epochs=10, batch_size=32, verbose=verbose)
    
    return(model_r, model_v)
```

```python
min_v = np.min(Y_v_train)
model_r, model_v = train(X_train, Y_r_train, Y_v_train, look_back, min_v, 1)
```

## IV) Empirical findings

```python
def test(X_test,Y_r_test, Y_v_test, model_r, model_v, scalers, rate, plot):
    pred_r = model_r.predict(X_test, verbose=0).reshape(-1)
    true_r = Y_r_test.reshape(-1)
    pred_v = model_v.predict(X_test, verbose=0).reshape(-1)
    true_v = Y_v_test.reshape(-1)
    min_v = np.min(true_v)
    x = np.arange(0,pred_v.shape[0],1)
    true_sr = (true_r - rate)/true_v
    
    pred_sr = (pred_r - rate)/pred_v
    error = (log_cosh(true_r, pred_r).numpy()*log_cosh(true_v, pred_v).numpy())**0.5
    #error = log_cosh(true_sr, pred_sr).numpy()
    #error = (np.mean((true_r - pred_r)**2)*np.mean((true_v - pred_v)**2))**0.5
    if plot:
        multiPlot(x,[[true_r,"true_r","-"],[pred_r,"pred_r","-"]],"days, days 0 : "+ str(test_start_date)+ " (test_start_date)","r","returns comparison")
        multiPlot(x,[[true_v,"true_v","-"],[pred_v,"pred_v","-"]],"days, days 0 : "+ str(test_start_date)+ " (test_start_date)","sigma","volatility comparison")
        multiPlot(x,[[true_sr,"true_sr","-"],[pred_sr,"pred_sr","-"]],"days, days 0 : "+ str(test_start_date)+ " (test_start_date)","sharp ratio","sharp ration comparison")
    return(error, pred_r, pred_v)
def reconstruct(X_test, pred_r, pred_v, horizon):
    h = horizon//2
    es_r = np.concatenate([pred_r[:h], pred_r, pred_r[:-h]])
    es_v = np.concatenate([pred_v[:h], pred_v, pred_v[:-h]])
    true_traj = scalers[0].inverse_transform(X_test[:,-1,0].reshape(-1,1))[:,0]
    n = true_traj.shape[0]
    reconstructed = np.zeros(n)
    for k in range(n):
        if k<h:
            reconstructed[k] = true_traj[0]
            for i in range(k):
                reconstructed[k] *=  np.random.normal(1+es_r[i], es_v[i], 1)[0]
        else:
            reconstructed[k] = true_traj[k-h]
            for i in range(horizon):
                reconstructed[k] *= np.random.normal(1+es_r[k-horizon+i], es_v[k-horizon+i], 1)[0]
    return(true_traj, reconstructed)
def compute_lead_lag(X,Y_r, Y_v, model_r, model_v, scalers, rate ,w_counterfactual, horizon):
    n_counterfactual = 7*w_counterfactual
    X_test,Y_r_test, Y_v_test, rate_test = X[train_duration-n_counterfactual:],Y_r[train_duration-n_counterfactual:], Y_v[train_duration-n_counterfactual:], rate[train_duration-n_counterfactual:]
    N = X_test.shape[0]//7
    ERROR = []
    lag = np.linspace(-n_counterfactual,-n_counterfactual+N*7-1, N)
    
    final_day = test_start_date + timedelta(days = - n_counterfactual + N*7-1)
    for k in range(N):
        error, pred_r, pred_v = test(X_test[k*7:(k+1)*7],Y_r_test[k*7:(k+1)*7], Y_v_test[k*7:(k+1)*7], model_r, model_v, scalers, rate_test[k*7:(k+1)*7] , False)
        ERROR.append(error)
    lag_train = lag[:w_counterfactual]
    lag_test = lag[w_counterfactual:]
    ERROR_train = ERROR[:w_counterfactual]
    ERROR_test = ERROR[w_counterfactual:]
    result_train = linregress(lag_train, ERROR_train)
    result_test = linregress(lag_test, ERROR_test)
    #multiPlot(lag,[[MSE,"MSE","."],[result_test.intercept + result_test.slope*lag," lagged horizons","-"],[result_train.intercept + result_train.slope*lag,"lead horizons (counterfactual)","-"]],"lags (weeks)","MSE","lead_lag MSE")
    plt.plot(lag_train, result_train.intercept + result_train.slope*lag_train, label='lead horizons (counterfactual) regression')
    plt.plot(lag_test, result_test.intercept + result_test.slope*lag_test, label='lagged horizons regression')
    plt.scatter(lag_train, ERROR_train, color='red', label='lead horizons (counterfactual)')
    plt.scatter(lag_test, ERROR_test, color='blue', label='lagged horizons')
    plt.xlabel('lags (weeks)')
    plt.ylabel('ERROR')
    plt.title('lead_lag ERROR')
    plt.legend()
    plt.show()
        


```

### IV.1) Prediction tests

```python
# Test on training data
error_train, pred_r, pred_v = test(X_train,Y_r_train, Y_v_train, model_r, model_v, scalers, rate[:X_train.shape[0]], True)
```

```python
#test on test data
error_test, pred_r, pred_v = test(X_test,Y_r_test, Y_v_test, model_r, model_v, scalers, rate[train_duration:], True)
```

```python
print("error on training data = " + str(error_train))
print("error on test data = " + str(error_test))
```

As expected, error on training data is smaller than on test data. The predictive performance of our model is rather good on the training data. But on the test data, the predictions quality is quite poor, the return and volatility curves do not overlap very well. In both cases, the sharp ratio : $(r-r_{\text{riskfree}})/\sigma$ overlap poorly du to the fact that the volatility is in the denominator so small prediction on small volatility would lead to big error in sharp ratio.

Let assume the index dynamics follow a Black&Scholes process with local return and volatility :

$dS_t = r_{loc} S_t dt + \sigma_{loc} S_t dB_t$

We can derive the formula in the discrete case :

$S_{n} = S_{n-1}+ S_{n-1}(\hat{r_{n-1}} + \hat{\sigma_{n-1}}N(0,1) ) $

Iteratively we deduce

$S_{n} = S_{n-7} + S_{n-7}\prod_{k=1}^{7}(\hat{r_{n-k}} + \hat{\sigma_{n-k}}N(0,1) ) $

We can reconstruct future index values over a horizon of 7 days by using past observations, predicted returns, and predicted volatility.

```python
true_traj, reconstructed = reconstruct(X_test, pred_r, pred_v, horizon)
multiPlot(np.arange(0, true_traj.shape[0], 1),[[true_traj,"true_traj","-"],[reconstructed,"reconstructed","-"]],"days","sigma","trajectory comparison")
```

This comparison gives a new interpretation on the prediction performance. We observe the reconstructed curve seems to be the true curve shifted to the right. This means the LSTM still tends to predict future values based on past trends, and therefore has difficulty anticipating sudden price changes. 

This is actually predictable, because the evolution of the S&P surely cannot be solely caused by ecological sentiments, here reduced to 2 indicators, and its past values.

The reconstruction is more noisy than original data, while the estimated volatility is not greater than the true volatility in average, this underlines the limits of our reconstruction model

It is expected to be so because in finance the validity of a model is limited in time depending on market paradigms. The more time passes, the further we move away from the market paradigms of the training period and the less precise the model trained in the old paradigm should be. In our particular case just by observing the values of the indices we see different paradigms: normal, the speculative bubble which is being built, the acceleration, the bursting...

## VI) Conclusion

The predictive performance of our model is rather good on the training data. But on the test data, the predictions quality is quite poor, the return and volatility curves do not overlap very well. It was believed that data from one-off events such as a speech by a central banker that strongly promotes sustainable development could be used to predict sudden changes in index trend, but despite this indicator, our LSTM still tends to predict future values based on past trends, and therefore has difficulty anticipating sudden price changes. 

Inaddition, prediction accuracy tends to decrease as we move away from the training period. This underlines that a model trained in this way is limited in time and that it is necessary to update it with recent data.

Our 2 indicators do not seem to improve the predictions. I think in order to improve the predictions it is surely appropriate to also include macro-economic and geopolitical data, especially for the energy sector (We all know that the post-covid recovery and the war in Ukraine has led to a surge in energy prices). Even if the results of this first exploration seem disappointing, we still encourage other teams to imagine and test other indicators to evaluate ecological themed market sentiment (like natural disaster occurence, the popularity of ecological policy measures...), and use them in addition to macroeconomic and geopolitical data.
